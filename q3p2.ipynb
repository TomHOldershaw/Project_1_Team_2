```python
import pandas as pd
```

```python
csv_path = "Resources/summer.csv"
```

```python
amjad_df=pd.read_csv(csv_path)
```

```python
amjad_df
```

```python
host_df=amjad_df.groupby('City')["Year"].unique()
host_df
```

```python
gbr_df=amjad_df.loc[(amjad_df["Country"] == "GBR") & ( (amjad_df["Year"] == 1896) | (amjad_df["Year"] == 1900) | (amjad_df["Year"] == 1904) )]
gbr_df
```

```python
athens_df=amjad_df.loc[(amjad_df["Country"] == "GRE") & ( (amjad_df["Year"] == 1896) | (amjad_df["Year"] == 1900) | (amjad_df["Year"] == 1904) )]
athens_df.head()
```

```python
medal_df=pd.DataFrame(amjad_df.groupby(['City','Country','Year'])['Medal'].value_counts())
medal_df.sort_values(by=["Year"],ascending=False, ignore_index=False)
medal_df.head(60)
```

```python
yearly_df=pd.DataFrame(amjad_df.groupby(['Country','Year'])['Medal'].value_counts().reset_index())
yearly_df
```
